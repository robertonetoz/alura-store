### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print(f"Faturamento Loja 1: R$ {faturamento_loja:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title('Faturamento por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(7,7))
plt.pie(faturamentos, labels=lojas, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'yellowgreen'])
plt.title('Participação no Faturamento Total por Loja')
plt.axis('equal')
plt.show()
```

# 2. Vendas por Categoria


```python
print("Loja 1 - Categorias mais vendidas:")
print(loja["Categoria do Produto"].value_counts(), "\n")

print("Loja 2 - Categorias mais vendidas:")
print(loja2["Categoria do Produto"].value_counts(), "\n")

print("Loja 3 - Categorias mais vendidas:")
print(loja3["Categoria do Produto"].value_counts(), "\n")

print("Loja 4 - Categorias mais vendidas:")
print(loja4["Categoria do Produto"].value_counts(), "\n")
```

```python
import matplotlib.pyplot as plt

categorias_loja = loja["Categoria do Produto"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    categorias_loja,
    labels=categorias_loja.index,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True
)
plt.title("Distribuição de Vendas por Categoria - Loja 1")
plt.axis("equal")
plt.show()

categorias_loja2 = loja2["Categoria do Produto"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    categorias_loja2,
    labels=categorias_loja.index,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True
)
plt.title("Distribuição de Vendas por Categoria - Loja 2")
plt.axis("equal")
plt.show()

categorias_loja3 = loja3["Categoria do Produto"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    categorias_loja3,
    labels=categorias_loja.index,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True
)
plt.title("Distribuição de Vendas por Categoria - Loja 3")
plt.axis("equal")
plt.show()

categorias_loja4 = loja4["Categoria do Produto"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    categorias_loja4,
    labels=categorias_loja.index,
    autopct="%1.1f%%",
    startangle=140,
    shadow=True
)
plt.title("Distribuição de Vendas por Categoria - Loja 4")
plt.axis("equal")
plt.show()
```

# 3. Média de Avaliação das Lojas

# Relatório Final

## **Introdução** <BR>
O objetivo desta análise foi apoiar o Senhor João em sua decisão sobre qual das suas quatro lojas deveria ser vendida. Para isso, foram analisados diversos aspectos de desempenho de cada loja: o faturamento total, a popularidade das categorias de produtos, a média de avaliação dos clientes, os produtos mais e menos vendidos e o custo médio de frete. Através dessas informações, foi possível identificar os pontos fortes e fracos de cada loja e recomendar, com base em dados concretos, a melhor opção para venda. <br><br>

## **Desenvolvimento** <br>
**💰 Faturamento Total**<br>
Loja 3 apresentou o maior faturamento entre as quatro, indicando um desempenho comercial forte.

Loja 2 teve o menor faturamento, sinalizando menor volume financeiro.

**📦 Categorias de Produtos Mais Vendidas**<br>
Em todas as lojas, móveis e eletrônicos foram as categorias mais populares.

Categorias como instrumentos musicais, livros e utilidades domésticas ficaram entre as menos vendidas.

**🌟 Avaliações dos Clientes**<br>
A Loja 4 teve a melhor média de avaliações dos clientes, indicando maior satisfação do consumidor.

A Loja 2 teve a pior média de avaliação, o que pode ser um indicativo de problemas de atendimento, logística ou qualidade dos produtos.

**🛒 Produtos Mais e Menos Vendidos**<br>
Produtos populares como cadeiras de escritório e smartphones apareceram como os mais vendidos em várias lojas.

Por outro lado, produtos de nicho, como harpas ou enciclopédias, estiveram entre os menos vendidos, impactando pouco no desempenho geral.

**🚚 Frete Médio**<br>
A Loja 4 também apresentou o menor custo médio de frete, um ponto positivo tanto do ponto de vista logístico quanto financeiro.

A Loja 1 teve o maior frete médio, o que pode reduzir sua competitividade.<br><br>

## **Conclusão e Recomendação**<br>
Após analisar todos os fatores, a recomendação é que o Senhor João venda a Loja 2.<br>

**Justificativa:**<br>

Apesar de ter custos de frete intermediários, a Loja 2 apresenta o menor faturamento e a pior média de avaliações entre todas.

Além disso, as vendas por categoria seguem a mesma tendência das demais lojas, mas sem se destacar em nenhuma frente relevante.

A loja tem menor rentabilidade e menor aceitação por parte dos clientes, o que reduz seu potencial de crescimento.

Em contraste, lojas como a Loja 3 (maior faturamento) e Loja 4 (melhores avaliações e menor frete) demonstram desempenho superior e devem ser mantidas como ativos principais.
